## PyTorch ~ Numpy   

Lets now see how similar and easy to learn pytorch is to numpy

**1. How to create a tensor(any n-dimensional matrix)**

```python
import torch
import numpy as np
```

```python
#PyTorch
t_tensor = torch.tensor([[2,3,4],[3,2,1]])

#Numpy
n_array = np.array([[2,3,4],[3,2,1]])

print('-------PyTorch------\n\n',t_tensor,'\n')
print('-------Numpy--------\n\n',n_array,'\n')
```

**2.Querying the shape of the torch tensor is simlar to numpy way with the keyword** --->  *torch_tensor.shape*

```python
#PyTorch
t_tensor = torch.tensor([[2,3,4],[3,2,1]])

#Numpy
n_array = np.array([[2,3,4],[3,2,1]])

print('-------PyTorch------\n\n',t_tensor.shape,'\n')
print('-------Numpy--------\n\n',n_array.shape,'\n')
```

**3.Random valued torch tensor can be created using** ---> *torch.rand(a, b) *

similar to the one in numpy as --->  np.random.rand(a, b)  where a & b are no.of rows and columns respectively

```python
#PyTorch
t_tensor = torch.rand(2,2)

#Numpy
n_array = np.random.rand(2,2)

print('-------PyTorch------\n\n',t_tensor,'\n')
print('-------Numpy--------\n\n',n_array,'\n')
```

**4.DOT product between two torch tensor is done with the help of** -----> *torch.matmul(tensor A, tensar B)*

Note: any matrix multiplication(dot product) will comply only if matrix dimension follows the heuristic (X x M) and (M x Y)

```python
#PyTorch
t_tensorA = torch.tensor([[2,3,4],[3,2,1]])
t_tensorB = torch.tensor([[1,2],[8,1],[2,5]])

#Numpy
n_arrayA = np.array([[2,3,4],[3,2,1]])
n_arrayB = np.array([[1,2],[8,1],[2,5]])

print('-------PyTorch------\n\n',torch.matmul(t_tensorA, t_tensorB),'\n')
print('-------Numpy--------\n\n',np.dot(n_arrayA, n_arrayB),'\n')
```

**5.Element Wise Multiplication between two torch tensor is done with the help of** -----> * tensor A * tensar B 

Note: This operation complies only if matrix dimension follows the heuristic (N x M) and (N x M) 

```python
#PyTorch
t_tensorA = torch.tensor([[2,3,4],[3,2,1]])
t_tensorB = torch.tensor([[1,2,8],[2,5,1]])

#Numpy
n_arrayA = np.array([[2,3,4],[3,2,1]])
n_arrayB = np.array([[1,2,8],[2,5,1]])

print('-------PyTorch------\n\n',t_tensorA * t_tensorB,'\n')
print('-------Numpy--------\n\n',np.multiply(n_arrayA, n_arrayB),'\n')
```

**5. Zeros, Ones and Idenity Torch tensor can be created using** * ---> torch.zeros(a, b); torch.ones(a, b); torch.eyes(a, b); repectively*

```python
#PyTorch
t_tensorA = torch.zeros(2,2)
t_tensorB = torch.ones(2,2)
t_tensorC = torch.eye(2,2)

#Numpy
n_arrayA = np.zeros([2,2])
n_arrayB = np.ones([2,2])
n_arrayC = np.identity(2)

print('-------PyTorch------\n\n',t_tensorA,'\n',t_tensorB, '\n', t_tensorC),
print('-------Numpy--------\n\n',n_arrayA,'\n', n_arrayB,'\n', n_arrayC)
```

**6.Convert numpy to torch tensor using** ---> * torch.from_numpy(numpy_array)*

**Convert troch to numpy using** ---> * torch_tensor.numpy()*

```python
#PyTorch
t_tensor = torch.rand(2,2)

#Numpy
n_array = np.random.rand(2,2)

print('-------Numpy array to PyTorch------\n\n',torch.from_numpy(n_array))
print('-------Torch to Numpy array--------\n\n',t_tensorA.numpy())
```
